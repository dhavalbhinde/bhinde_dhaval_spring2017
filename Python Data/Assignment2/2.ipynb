```python
# Part 1
import os, json
import shutil

# Get all JSON files from the root folder named 'Data'
path_to_json = 'Data'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]

for js in json_files:
    with open(os.path.join(path_to_json, js)) as json_file:
        data = json.load(json_file)
        country = (data['location']['country'])    # get country from JSON
        state = (data['location']['state'])        # get state from JSON
        city = (data['location']['city'])          # get city from JSON
        term = (data['term'])                      # get term from JSON
        
        destpath = 'Processed_Data/'+country+'/'+state+'/'+city+'/'+term  # build destination path hierarchy
        srcfile = 'Data/'+js                                              # src file path
        
        assert not os.path.isabs(srcfile)
        dstdir =  destpath
        if not os.path.exists(dstdir):  # create all directories, if it does not exists already
               os.makedirs(dstdir)
        shutil.copy(srcfile, dstdir)    # copy command

print("Execution Complete!")
        
```

```python
# Part2
import os, json
import shutil
import csv

# Create file data_restaurants and add the title row

with open('data_restaurants.csv','w',newline='') as out:
    csv_out=csv.writer(out)
    csv_out.writerow(['Name of restaurant','City','CountryCode','Day of Week','Start Time HR','Start Time Min','End Time HR','End Time Min'])
    
# Read JSON files with term 'restaurants' and fetch the required data
# Write data one row at a time into the CSV file.
# Use try catch exception for all the JSON files which donot have open hrs information

    for js in json_files:
        with open(os.path.join(path_to_json, js)) as json_file:
            data = json.load(json_file)
            term = (data['term'])
            name = (data['name'])
            city = (data['location']['city'])
            country = (data['location']['country'])
            if term == 'restaurants':
                try:
                    for open_time in data['hours']:
                        for day in open_time['open']:
                            day_out = (day['day'])
                            start_time = (day['start'])
                            start_time_hour = start_time[:2]
                            start_time_min  = start_time[-2:]
                            end_time = (day['end'])
                            end_time_hour = end_time[:2]
                            end_time_min  = end_time[-2:]
                            csv_out.writerow([name,city,country,day_out,start_time_hour,start_time_min,end_time_hour,end_time_min])
                except Exception:
                    print("Exception for name:"+name+", :open hrs not present")        
                    
```
